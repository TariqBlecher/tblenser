```python
def lensing_interaction(zsrc=1, sigma_arcsec=1, x_off=0, y_off=0):
    abell_370.widget_sim(zsrc, sigma_arcsec, x_off, y_off)
    image = pf.getdata(abell_370.prefix+'_image.fits')
    source = pf.getdata(abell_370.prefix+'_source.fits')
    imshow(image-source,origin='lower',extent = [-30,30,-30,30])
    colorbar()
    show()
    
    
interact(lensing_interaction,zsrc=(0.4,2.0,0.1),sigma_arcsec=(0.2,2.0,0.1),
         x_off=(-20,40,1),y_off=(-20,60,1));
```
